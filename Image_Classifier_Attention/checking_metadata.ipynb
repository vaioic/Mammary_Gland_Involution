# Verify that metadata file is set up correctly

```python
import pandas as pd
import numpy as np
```

```python
metadata = pd.read_csv("2026.1.11.AnnotationlevelKey.csv",index_col=0)
```

```python
metadata.head()
```

```python
wt_metadata = metadata[metadata['Genotype']=='WT']
wt_metadata['Timepoint'].unique()
```

```python
conditions = [
    (wt_metadata['Timepoint'] == 'Nulli'),
    (wt_metadata['Timepoint'] == 'IV4'),
    (wt_metadata['Timepoint'] == 'Pregnant'),
    (wt_metadata['Timepoint'] == 'IV6'),
    (wt_metadata['Timepoint'] == 'IV14'),
    (wt_metadata['Timepoint'] == 'IV10'),
    (wt_metadata['Timepoint'] == 'IV28'),
    (wt_metadata['Timepoint'] == 'IV8'),
    (wt_metadata['Timepoint'] == 'IV1'),
    (wt_metadata['Timepoint'] == 'IV2'),
    (wt_metadata['Timepoint'] == 'Lactating')
]

labels = list(range(0,11))
```

```python
print(labels)
```

```python
wt_metadata['label'] = np.select(conditions,labels)
wt_metadata.head()
```

```python
wt_metadata.to_csv('metadata.csv')
```

```python
subset_df = wt_metadata.sample(n=20)
subset_df.to_csv('metadata_subset.csv')
```
